```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import csv

# 웹 드라이버 설정
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# 웹 페이지 로드
url = "http://kpat.kipris.or.kr/kpat/searchLogina.do?next=MainSearch#page1"
driver.get(url)

# WebDriverWait 객체 설정
wait = WebDriverWait(driver, 3)

# 빈 리스트 생성
company_list = []
search_result_list = []

# 이전에 검색된 결과를 저장하기 위한 세트
previous_results = set()

# CSV 파일 읽기
with open('./top200_name_list.csv', 'r', encoding='utf-8') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        data = row[0]

        search_input = driver.find_element(By.ID, 'queryText')
        search_input.clear()
        search_input.send_keys("*AP=[" + data + "]")
        search_input.send_keys(Keys.ENTER)

        print("회사명:", data)

        try:
            # CSS 선택자를 사용하여 검색 결과 가져오기
            mainsearch_info_list = wait.until(EC.visibility_of_all_elements_located((By.CSS_SELECTOR, '#mainsearch_info_list > div.mainlist_topinfo > li:nth-child(3) > span.point01 > a')))
            search_results_temp = [tag.text for tag in mainsearch_info_list]
        except Exception as e1:
            print("CSS 선택자로 검색 결과를 찾는 동안 예외 발생:", e1)
            try:
                # XPath를 사용하여 검색 결과 가져오기
                mainsearch_info_list = wait.until(EC.visibility_of_all_elements_located((By.XPATH, '//*[@id="mainsearch_info_list"]/div[1]/li[2]/span[2]/a')))
                search_results_temp = [tag.text for tag in mainsearch_info_list]
            except Exception as e2:
                print("XPath로 검색 결과를 찾는 동안 예외 발생:", e2)
                search_results_temp = ["검색 결과가 없습니다."]

        # 중복 제거 로직 추가
        search_results = []
        for result in search_results_temp:
            if result not in previous_results:
                search_results.append(result)
                previous_results.add(result)

        company_list.append(data)
        search_result_list.append(search_results)

# 데이터 프레임 생성
df = pd.DataFrame({'회사명': company_list, '검색 결과': search_result_list})

# 결과를 CSV 파일로 저장
df.to_csv('검색결과3.csv', index=False)

# 웹 드라이버 종료
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import csv

# 웹 드라이버 설정
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# 웹 페이지 로드
url = "http://kpat.kipris.or.kr/kpat/searchLogina.do?next=MainSearch#page1"
driver.get(url)

# WebDriverWait 객체 설정 (대기 시간을 늘림)
wait = WebDriverWait(driver, 10)

# 빈 리스트 생성
company_list = []
search_result_list = []

# 이전에 검색된 결과를 저장하기 위한 세트
previous_results = set()

# CSV 파일 읽기
with open('./top200_name_list.csv', 'r', encoding='utf-8') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        data = row[0]

        search_input = driver.find_element(By.ID, 'queryText')
        search_input.clear()
        search_input.send_keys("*AP=[" + data + "]")
        search_input.send_keys(Keys.ENTER)

        print("회사명:", data)

        try:
            # 검색 결과를 가져오기 위해 presence_of_all_elements_located 사용
            mainsearch_info_list = wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, '#mainsearch_info_list > div.mainlist_topinfo > li:nth-child(3) > span.point01 > a')))
            search_results_temp = [tag.text for tag in mainsearch_info_list]

            # 중복 제거 로직 추가
            search_results = []
            for result in search_results_temp:
                if result not in previous_results:
                    search_results.append(result)
                    previous_results.add(result)

            company_list.append(data)
            search_result_list.append(search_results)

        except Exception as e:
            print("검색 결과를 가져오는 동안 예외 발생:", e)
            search_results_temp = ["검색 결과가 없습니다."]
            company_list.append(data)
            search_result_list.append(search_results_temp)

# 데이터 프레임 생성
df = pd.DataFrame({'회사명': company_list, '검색 결과': search_result_list})

# 결과를 CSV 파일로 저장
df.to_csv('검색결과6.csv', index=False)

# 웹 드라이버 종료
driver.quit()
```
