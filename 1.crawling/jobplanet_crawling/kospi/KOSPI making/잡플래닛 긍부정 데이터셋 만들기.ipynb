```python
import pandas as pd 
```

```python
gb_df = pd.read_csv('./KOSPI result/kospi_date_pros_cons_jobplanet_result.csv')
```

```python
gb_df
```

```python
문장 = []
긍정 = []
부정 = []
for good, bad in zip(gb_df['장점 리뷰'], gb_df['단점 리뷰']) :
    문장.append(str(good))
    문장.append(str(bad))
    긍정.append(1)
    긍정.append(0)
    부정.append(0)
    부정.append(1)
```

```python
label_gb_df = pd.DataFrame({'문장' : 문장, '긍정' : 긍정, '부정' : 부정})
```

```python
label_gb_df
```

```python
label_gb_df = label_gb_df[label_gb_df['문장'].apply(lambda x: len(x) > 30)]
label_gb_df = label_gb_df[label_gb_df['문장'].apply(lambda x: len(x) < 200)]
```

```python
label_gb_df.to_csv('./jobplanet_긍부정데이터셋')
```
