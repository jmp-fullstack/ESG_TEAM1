```python
# 크롤링데이터 가져오기
import pandas as pd
df = pd.read_csv('지속가능경영보고서전처리전내용.csv')
df
```

```python
data = []
for i in df['내용']:
    d = eval(i)
    data.extend(d)
data
```

```python
# # 문장만 모아 텍스트 파일로 저장
# with open('지속가능경영보고서문장.txt', 'w') as f:
#     f.write('\n'.join(data))

# # 저장한 파일 가져오기
# with open('지속가능경영보고서문장.txt', 'r') as f:
#     data = f.readlines()
```

```python
# 전처리
import re
pattern = r"[^ㄱ-ㅎㅏ-ㅣ가-힣., ]"
cleaned_data = [re.sub(pattern, '', sentence) for sentence in data]
cleaned_data
```

```python
# # 전처리된 데이터를 텍스트 파일에 쓰기
# with open('지속가능경영보고서전처리완료.txt', 'w', encoding='utf-8') as file:
#     for sentence in cleaned_data:
#         file.write(sentence + '\n')
```

```python
# 명사만 추출
from konlpy.tag import Mecab

def extract_nouns(sentences):
    mecab = Mecab(dicpath="C:\mecab\mecab-ko-dic")
    nouns_list = []
    
    for sentence in sentences:
        # 형태소 분석 수행
        pos_tags = mecab.pos(sentence)
        
        # 명사 추출 (한 글자가 아닌 경우에만 리스트에 추가)
        nouns = [word for word, pos in pos_tags if pos.startswith('N') and len(word) > 1]
        
        # 추출된 명사가 있을 경우에만 리스트에 추가
        if nouns:
            nouns_list.append(nouns)
    
    return nouns_list

# 명사 추출
nouns_list = extract_nouns(cleaned_data)
nouns_list
```

```python
# Word2Vec으로 토큰화된 데이터 학습
from gensim.models import Word2Vec

model = Word2Vec(sentences = nouns_list, vector_size = 100, window = 5, min_count = 5, workers = 4, sg = 0)

# 완성된 임베딩 매트릭스의 크기 확인
model.wv.vectors.shape
```

```python
# 주제별 키워드 가져오기
f = open("E_key_words.txt", "r")
E_key_words = [word.strip() for word in f.readlines()]
f.close()

f = open("S_key_words.txt", "r")
S_key_words = [word.strip() for word in f.readlines()]
f.close()

f = open("G_key_words.txt", "r")
G_key_words = [word.strip() for word in f.readlines()]
f.close()
```

```python
# E 키워드 추출
e_list = []
for word in E_key_words:
    try:
        e = model.wv.most_similar(word, topn=30)
        for i in e:
            if i[1]>=0.5:
                e_list.append(i[0])
    except:
        print('[[', word, ']]는 지속가능경영보고서로부터 추출된 단어모음집에 존재하지 않습니다.')

E_LIST = []
for i in set(e_list):
    E_LIST.append((i, 'E'))
E_LIST
```

```python
# S 키워드 추출
s_list = []
for word in S_key_words:
    try:
        s = model.wv.most_similar(word, topn=30)
        for i in s:
            if i[1]>=0.5:
                s_list.append(i[0])
    except:
        print('[[', word, ']]는 지속가능경영보고서로부터 추출된 단어모음집에 존재하지 않습니다.')

S_LIST = []
for i in set(s_list):
    S_LIST.append((i, 'S'))
S_LIST
```

```python
# G 키워드 추출
g_list = []
for word in G_key_words:
    try:
        g = model.wv.most_similar(word, topn=30)
        for i in g:
            if i[1]>=0.5:
                g_list.append(i[0])
    except:
        print('[[', word, ']]는 지속가능경영보고서로부터 추출된 단어모음집에 존재하지 않습니다.')

G_LIST = []
for i in set(g_list):
    G_LIST.append((i, 'G'))
G_LIST
```

```python
# E, S, G 키워드 저장
ESG_LIST = E_LIST + S_LIST + G_LIST

col1 = "keyword", "label"

with open('지속가능경영보고서_키워드.csv', 'w') as file:
    file.write("keyword, label\n") # column 
    for item in ESG_LIST:
        file.write(f"{item[0]}, {item[1]}\n") # value
```
