```python
import pandas as pd
```

```python
green_df = pd.read_csv('green_final.csv')
```

```python
green_df['기업명'].tolist()
```

```python
green_df_count = green_df.groupby(['기업명', '인증연도']).size().reset_index(name='개수')
green_df_count
```

```python
green_df_count['개수'].unique()
```

```python
import matplotlib.pyplot as plt

plt.plot(green_df_count['개수'].tolist())
plt.show()
```

```python
# 개수에 따라 점수 계산하여 점수 열 추가
def calculate_score(count):
    if count == 0:
        return 0
    elif 1 <= count <= 2:
        return 3
    else:
        return 5

green_df_count["점수"] = green_df_count["개수"].apply(calculate_score)
green_df_count
```

```python
green_df_count.to_csv('녹색점수.csv',index=False)
```

```python
rc_df = pd.read_csv('recall_final.csv')
```

```python
rc_df[:3]
```

```python
rc_df_count = rc_df.groupby(['기업명', '인증연도', '리콜종류']).size().reset_index(name='개수')
rc_df_count
```

```python
import matplotlib.pyplot as plt

plt.plot(rc_df_count['개수'].tolist())
plt.show()
```

```python
# 개수에 따라 점수 계산하여 점수 열 추가
def calculate_score(count):
    if count == 0:
        return 5
    elif count == 1:
        return 4
    elif count == 2:
        return 3
    elif count == 3:
        return 2
    elif count == 4:
        return 1
    else:
        return 0

rc_df_count["점수"] = rc_df_count["개수"].apply(calculate_score)
rc_df_count
```

```python
rc_df_count.to_csv('리콜점수.csv',index=False)
```
