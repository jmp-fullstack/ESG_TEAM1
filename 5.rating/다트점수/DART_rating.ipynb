```python
import pandas as pd
```

```python
# 전처리 완료된 파일 가져오기
# rating_dart = pd.read_excel('다트전처리완료_2023.xlsx')
rating_dart = pd.read_excel('다트전처리완료_2022.xlsx')
```

```python
# 사용할 열만 선택
rating_dart = rating_dart[['회사명', '업종', '최대주주_기말주식지분율', '감사의견']].copy()
rating_dart
```

```python
# 점수 열 추가 및 초기값 0 설정
rating_dart['최대주주_기말주식지분율_점수'] = 0
rating_dart['감사의견_점수'] = 0
```

```python
# 최대주주_기말주식비율이 60 이상인 경우 1.5점, 정보가 없는 경우 3.75점, 정보가 있으면서 60 미만인 경우 7.5점 (7.5점 만점)
rating_dart.loc[rating_dart['최대주주_기말주식지분율']>=60, '최대주주_기말주식지분율_점수'] =1.5
rating_dart.loc[rating_dart['최대주주_기말주식지분율'].isnull(), '최대주주_기말주식지분율_점수'] =3.75
rating_dart.loc[(rating_dart['최대주주_기말주식지분율'].notnull()) & (rating_dart['최대주주_기말주식지분율'] < 60), '최대주주_기말주식지분율_점수'] = 7.5

# 감사의견에 적정이라는 값이 들어가 있지 않거나 값이 없는 경우는 0점, 적정이 들어간 경우는 7.5점
rating_dart.loc[rating_dart['감사의견'].str.contains('적정', na=False), '감사의견_점수'] = 7.5

# 점수 총합
rating_dart['총점수'] = rating_dart['최대주주_기말주식지분율_점수'] + rating_dart['감사의견_점수']

rating_dart
```

```python
# 점수 포함된 파일 저장
# rating_dart.to_excel('rating_dart_2023.xlsx', index=False)
rating_dart.to_excel('rating_dart_2022.xlsx', index=False)
```

```python
# 파일 불러오기 (확인용)
# pd.read_excel('rating_dart_2023.xlsx')
pd.read_excel('rating_dart_2022.xlsx')
```

자료 모아서 정리

```python
import pandas as pd
dart_2023 = pd.read_excel('rating_dart_2023.xlsx')
dart_2022 = pd.read_excel('rating_dart_2022.xlsx')
```

```python
# 년도 열 추가
dart_2023['년도'] = 2023
dart_2022['년도'] = 2022
```

```python
df = pd.concat([dart_2023,dart_2022],axis=0)

# 회사명을 기준으로 정렬
df.sort_values(by='회사명', inplace=True)
```

```python
# 열 순서 변경
df = df.reindex(columns=['회사명', '업종', '년도', '최대주주_기말주식지분율', '감사의견' '최대주주_기말주식지분율_점수', '감사의견_점수', '총점수'])
df
```

```python
# 저장
df.to_excel('rating_dart_total.xlsx', index=False)
```
